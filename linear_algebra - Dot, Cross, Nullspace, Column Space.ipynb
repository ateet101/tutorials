# Vector Dot Product and Cross Products

## Theory:
Dot product of two vectors $a*b = 0$ if the angle between them is 90

Dot product of two vectors $a*b = ||a||*||b||*cos(\Theta)$


$cos(\Theta) = \frac{a*b}{|a|*|b|}$ where $|a| = \sqrt{a_1^2 + a_2^2 + \dots + a_n^2}$

$Proj_b(a) = \frac{a*b}{b^2}*b$ is the Projection of vector a onto vector b

Dot product of a vector with itself is $a*a =||a||^2$

$cos(\Theta)=\frac{adj}{hyp}$

$cos(\Theta)=\frac{adj}{||a||}$

$cos = ||a||sin\Theta$

Let's take an example of work where W=f*d - Work equals force times D.
Let D stay on the x-axis and f also stay on the x-axis.  The work put in would simply be force times distance

However, if force was applied at an inclined angle, more work would have to be put in to cover the same distance. Therefore, the dot product of force and distance is the projection of the force vector onto the distance vector

```python
f = np.array([3,4]).transpose() #The transpose is because a 1x2 vector has to be dotted with a 2x1 vector
d = np.array([9,0])
W = np.dot(f,d)
print("The dot product of the Force vector (3,4) times \nthe distance vector (9,0) is Work =\n", W)


C = CoordSysCartesian('C')
i,j,k = C.base_vectors()#i,j,k are for the x,y,z axis
v1 = 9*i + 0*j#specifies coordinates  of (9,0,0)
v2 = 3*i + 4*j
v11 = v1.to_matrix(C)
v22 = v2.to_matrix(C)
v3 = (v1.projection(v2)).to_matrix(C) #Projection of v2 onto v1
plt.plot([0,v22[0]],[0,v22[1]], color='b', label= 'Force')
plt.plot([0,v11[0]],[0,v11[1]], color='r', label= 'Distance')
plt.plot([0,v3[0]],[0,v3[1]], color='g', label= 'Work: Projection of Force on Distance')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('Dot Product of Force times distance')
plt.show()
print("The projection of the Force vector (3,4) onto the distance vector (9,0) =\n", v3[0])
C = CoordSysCartesian('C')
i,j,k = C.base_vectors()
v1 = 2*i + 1*j
v2 = 2*i + 3*j
v3 = (v1.projection(v2)).to_matrix(C) #projection of v2 onto v1
```

Is Ax a subspace?

1) $A\bar 0 = 0$; contains 0 vector $\bar 0 \epsilon N$

2) $A\bar v_1 = \bar 0$ and $A\bar v_2 = \bar 0$

$A(\bar v_1+\bar v_2) = A\bar v_1 + A\bar v_2$

$\bar v_1 + \bar v_2 \epsilon N$ closed under addition

3)$A(c\bar v_1) = cA\bar v_1 = c\bar0 = \bar0$ closed under multiplication

Therefore N is Nullspace of A $N=N(A)$

$N(A) = \bar x \epsilon R^n | A\bar x = \bar 0$ The null space of A(N(A)) is the set of all vectors(x) that are a member of $R^n$ such that Matrix A times the vector is equal to the 0 vector

N(A) = N(rref(A))
