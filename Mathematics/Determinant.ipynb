## Determinant

Determinant - the scaled area by two vectors or a matrix 

Determinant of 2 vectors in 2D Space
\begin{matrix}
det
\begin{bmatrix}
a&b\\
c&d
\end{bmatrix}
=
\begin{pmatrix}
a*d + b*c\\
\end{pmatrix}
\end{matrix}

Determinant of 3 vectors in 3D Space
\begin{matrix}
det
\begin{bmatrix}
i&j&k\\
a&b&c\\
d&e&f\\
g&h&i
\end{bmatrix}
=
\begin{pmatrix}
i(e*i-f*h)-j(a*i-c*g)+k(a*e-b*d)
\end{pmatrix}
\end{matrix}

Let's take 2 vectors $\bar i$ $\bar j$ and find the determinant

\begin{matrix}
\textrm{Original Coordinates}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
0&2\\
-1.5&1
\end{bmatrix}
= (0*1) - (-1.5*2) = 3
\end{matrix}

The determinant is 0 if the area is 0 meaning that the two vectors are on the same line(2D to 1D)
\begin{matrix}
\textrm{Zero Determinant}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
4&2\\
2&1
\end{bmatrix}
= (4*1) - (2*2) = 0
\end{matrix}

The determinant is negative when the matrix is inverted or flipped
\begin{matrix}
\textrm{Negative Determinant}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
2&1\\
-1&-3
\end{bmatrix}
= (2*-3) - (-1*1) = -5
\end{matrix}

```python
vector1 = Matrix(([0,2],[-1.5,1]))
vector2 = Matrix(([4,2],[2,1]))
vector3 = Matrix(([2,1],[-1,-3]))
det_vector1 = vector1.det()

fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.plot([0,vector1[0,0]], [0,vector1[1,0]], 'r-')
ax1.plot([0,vector1[0,1]], [0,vector1[1,1]], 'b-')
ax1.set_xlim(-3,3)
ax1.set_ylim(-3,3)
ax1.set_title('Determinant or Area = 3')
ax1.fill_between([0,2], [-1.5,-1.5], [0,1], color='grey', alpha='0.5')

ax2 = fig.add_subplot(222)
ax2.plot([0,vector2[0,0]], [0,vector2[1,0]], 'r-')
ax2.plot([0,vector2[0,1]], [0,vector2[1,1]], 'b-')
ax2.set_xlim(-3,3)
ax2.set_ylim(-3,3)
ax2.set_title('Determinant or Area = 0')

ax3 = fig.add_subplot(223)
ax3.plot([0,vector3[0,0]], [0,vector3[1,0]], 'r-')
ax3.plot([0,vector3[0,1]], [0,vector3[1,1]], 'b-')
ax3.set_xlim(-3,3)
ax3.set_ylim(-3,3)
ax3.set_title('Determinant or Area = -5')
ax3.fill_between([1,2], [-3,-1], [-.5,-1],color='grey', alpha='0.5')
ax3.fill_between([0,1], [0,-3], [0,-.5],color='grey', alpha='0.5') 
#ax3.fill_between([x1,x2], [y1_lower,y2_lower], [y1_higher,y2_higher])

ax4 = fig.add_subplot(224)
#ax4.plot([1,2,3,4], [0,0,1,1], 'g-')


plt.tight_layout()
fig = plt.gcf()
plt.show()
```

The coordinates of a new vector (x,y) can be computed by multiplying the basis vectors i, j in Matrix form to the new vector

\begin{matrix}
\textrm{Original Coordinates}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
1x+0y\\
0x+1y
\end{bmatrix}
\end{matrix}

Any vector (x,y) can be transformed into a new matrix by multiplying the vector by the rotation matrixx, the shear matrix, or both.  The following matrix means that the rotation matrix is being applied first and the shear matrix is being applied seconde
\begin{matrix}
\begin{bmatrix}

## The Determinant in a 3D Space is a Volume of a Cube
The dot product of a vector and (x,y,z) is equal to the det of 2 other vectors in 3D space
\begin{matrix}
\begin{bmatrix}
p_1\\
p_2\\
p_3
\end{bmatrix}
\begin{bmatrix}
x\\
y\\
z
\end{bmatrix}
= det
\begin{bmatrix}
x&v_1&w_1\\
y&v_2&w_2\\
z&v_3&w_3\\
\end{bmatrix}
\end{matrix}
\begin{pmatrix}
p_1 = v_2w_3-v_3w_2\\
p_1 = v_3w_1-v_1w_3\\
p_1 = v_1w_2-v_2w_1\\
\end{pmatrix}


```python
v1 = np.array([1,0,0])
v2 = np.array([0,1,0])
v3 = np.cross(v1,v2)
fig1 = plt.figure()
ax = fig1.add_subplot(111, projection='3d')

ax.plot([0,v1[0]],[0,v1[1]],[0,v1[2]], color = 'r', label = 'Vector 1')
ax.plot([0,v2[0]],[0,v2[1]],[0,v2[2]], color = 'b', label = 'Vector 2')
ax.plot([0,v3[0]],[0,v3[1]],[0,v3[2]], color = 'g', label = 'Vector 1 X Vector 2')

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.legend()

r = [0,1]
X, Y = np.meshgrid(r, r)
ax.plot_surface(X,Y,1, alpha=0.5)
ax.plot_surface(X,Y,0, alpha=0.5)
ax.plot_surface(X,0,Y, alpha=0.5)
ax.plot_surface(X,1,Y, alpha=0.5)
ax.plot_surface(1,X,Y, alpha=0.5)
ax.plot_surface(0,X,Y, alpha=0.5)
ax.set_xlim(-.5,1.5)
ax.set_ylim(-.5,1.5)
ax.set_zlim(-.5,1.5)

plt.show()
```
