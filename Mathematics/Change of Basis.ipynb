## EigenValues
EigenVectors remain on its own span or 1 Dimension
EigenValue - increases the magnitude of the vector

EigenValue ~ $\lambda$
$$A\bar v = \lambda \bar v$$
\begin{matrix}
\lambda
\begin{bmatrix}
1&0&0\\
0&1&0\\
0&0&1
\end{bmatrix}
=
\begin{bmatrix}
1-\lambda &0&0\\
0&1-\lambda &0\\
0&0&1-\lambda 
\end{bmatrix}
\end{matrix}

$$A\bar v = (\lambda I)\bar v$$

$$(A\bar v - \lambda I)\bar v=0$$

$$det(A\bar v - \lambda I)\bar =0$$

\begin{matrix}
\begin{bmatrix}
3&1\\
0&2
\end{bmatrix}
-> det
\begin{bmatrix}
3-\lambda &1\\
0&2-\lambda\\
\end{bmatrix}
=(3-\lambda)(2-\lambda)-1*0 -> \lambda = 2,3
\end{matrix}

\begin{matrix}
\begin{bmatrix}
0&-1\\
1&0
\end{bmatrix}
-> det
\begin{bmatrix}
-\lambda &-1\\
1&-\lambda\\
\end{bmatrix}
=(-\lambda)(-\lambda)-1*1 = \lambda^2 + 1 = 0 -> \lambda = i,-i
\end{matrix}
