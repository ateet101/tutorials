## Linear System of Equations

Formal Linearity Properties
$$L(\bar v + \bar w) = L(\bar v) + L(\bar w)$$
$$L(c\bar v) = cL(\bar v)$$

\begin{matrix}
\begin{bmatrix}
    a_1\\
    a_2\\
    a_3\\
\end{bmatrix}
\begin{bmatrix}
    b_1\\
    b_2\\
    b_3\\
\end{bmatrix}
\begin{bmatrix}
    c_1\\
    c_2\\
    c_3\\
\end{bmatrix}=
\begin{bmatrix}
    v_1\\
    v_2\\
    v_3\\
\end{bmatrix}
\end{matrix}
$A\bar x = \bar v$

Solving for a vector $\bar x$ where after applying a transformation A, $\bar x$ lies on vector $\bar v$

If the det(A) does not equal 0, there will always be an x that lies on v
If the det(A) = 0, there 2D -> 1D. We no longer have an area in 2D space

\begin{matrix}
\begin{bmatrix}
    a_1&b_1&c_1\\
    a_2&b_2&c_2\\
    a_3&b_3&c_3\\
\end{bmatrix}
\begin{bmatrix}
    x\\
    y\\
    z\\
\end{bmatrix}
=
\begin{bmatrix}
    v_1\\
    v_2\\
    v_3\\
\end{bmatrix}
\end{matrix}

\begin{equation}
a_1x + b_1y + c_1z = v_1\\
a_2x + b_2y + c_2z = v_2\\
a_3x + b_3y + c_3z = v_3\\
\end{equation}

$\bar x = A^{-1}\bar v$ gives you v that lies on x

## Inverse
The inverse of a Matrix $A^{-1}$ implies that A is the opposite of $A^{-1}$
If A was a matrix 90 Counter Clockwise, then $A^{-1}$ would be 90 Clockwise

\begin{matrix}
A = 
\begin{bmatrix}
    0&-1\\
    1&0
\end{bmatrix}
A^{-1} =
\begin{bmatrix}
    0&1\\
    -1&0
\end{bmatrix}
A^{-1}A =
\begin{bmatrix}
    1&0\\
    0&1
\end{bmatrix}
\end{matrix}

```python
A = Matrix(([0,-1],[1,0]))
A_inv = A.inv()#Matrix(([0,1],[-1,0]))
iden = A_inv*A
fig = plt.figure()

ax1 = fig.add_subplot(311)
ax1.plot([0, A[0,0]], [0,A[1,0]], 'r-')
ax1.plot([0,A[0,1]], [0,A[1,1]], 'b-')
ax1.set_xlim(-3,3)
ax1.set_ylim(-3,3)
ax1.set_title('A')

ax2 = fig.add_subplot(312)
ax2.plot([0,A_inv[0,0]], [0,A_inv[1,0]], 'r-')
ax2.plot([0,A_inv[0,1]], [0,A_inv[1,1]], 'b-')
ax2.set_xlim(-3,3)
ax2.set_ylim(-3,3)
ax2.set_title('Inverse of A')

ax3 = fig.add_subplot(313)
ax3.plot([0,iden[0,0]], [0,iden[1,0]], 'r-')
ax3.plot([0,iden[0,1]], [0,iden[1,1]], 'b-')
ax3.set_xlim(-3,3)
ax3.set_ylim(-3,3)
ax3.set_title('Inverse of A * A or Identity Matrix')

plt.tight_layout()
fig = plt.gcf()
plt.show()
```

Composition Matrix of a rotation and a shear matrix is both multiplied together
\begin{matrix}
\begin{bmatrix}
1 & 1\\
0&1
\end{bmatrix}
\begin{bmatrix}
0&-1\\
1&0
\end{bmatrix}
=

\end{bmatrix}
\end{matrix}

\begin{matrix}
\begin{bmatrix}
a&b\\
c&d
\end{bmatrix}
\begin{bmatrix}
e&f\\
g&h
\end{bmatrix}
=
\begin{bmatrix}
a*e + b*g & a*f + b*h\\
c*e + d*g & c*f + d*h
\end{bmatrix}
\end{matrix}

Let's take 2 vectors $\bar u$ $\bar v$ and place it on a x, y coordinate system

\begin{matrix}
\textrm{Original Coordinates}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}
\end{matrix}

A shear matrix is a addition of a multiple of one row or one column to another
\begin{matrix}
\textrm{Shear Coordinates}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
1&1\\
0&1
\end{bmatrix}
\end{matrix}

A 90 degree rotation matrix rotates the $\bar j$ vector from coordinates [1,0] to [-1,0]
\begin{matrix}
\textrm{Rotation Matrix}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
0&-1\\
1&0
\end{bmatrix}
\end{matrix}

The coordinates of a new vector (x,y) can be computed by multiplying the basis vectors i, j in Matrix form to the new vector

\begin{matrix}
\textrm{Original Coordinates}
\begin{bmatrix}
i&j
\end{bmatrix}
=
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
1x+0y\\
0x+1y
\end{bmatrix}
\end{matrix}

Any vector (x,y) can be transformed into a new matrix by multiplying the vector by the rotation matrixx, the shear matrix, or both.  The following matrix means that the rotation matrix is being applied first and the shear matrix is being applied seconde
\begin{matrix}
\begin{bmatrix}
1 & 1\\
0&1
\end{bmatrix}
\begin{pmatrix}
\begin{bmatrix}
0&-1\\
1&0
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
\end{pmatrix}
=
\begin{bmatrix}
1&-1\\
1&0
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
\end{matrix}




```python
orig_matrix = Matrix(([1,0],[0,1]))
shear_matrix = Matrix(([1,1],[0,1]))
rot_matrix = Matrix(([-1,0],[0,1]))
com_matrix = shear_matrix *(rot_matrix)#Matrix(([1,-1],[1,0]))
vector = Matrix(([2,3]))

rot_vector = rot_matrix*(vector)#rot_vector = rot_matrix.dot(vector)
shear_vector = shear_matrix*(vector) #shear_vector = shear_matrix.dot(vector)
shear_rot_vector = shear_matrix*(rot_matrix*vector)#shear_rot_vector = shear_matrix.dot(rot_matrix.dot(vector))
com_vector = com_matrix*(vector)#com_vector = com_matrix.dot(vector)

fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.plot([0,orig_matrix[0,0]], [0,orig_matrix[1,0]], 'r-')
ax1.plot([0,orig_matrix[0,1]], [0,orig_matrix[1,1]], 'b-')
ax1.plot([0,vector[0]], [0,vector[1]], 'g-')
ax1.set_xlim(-3,3)
ax1.set_ylim(-3,3)
ax1.set_title('Original Matrix')

ax2 = fig.add_subplot(222)
ax2.plot([0,shear_matrix[0,0]], [0,shear_matrix[1,0]], 'r-')
ax2.plot([0,shear_matrix[0,1]], [0,shear_matrix[1,1]], 'b-')
ax2.plot([0,shear_vector[0]], [0,shear_vector[1]], 'g-')
ax2.set_xlim(-3,3)
ax2.set_ylim(-3,3)
ax2.set_title('Shear Matrix')

ax3 = fig.add_subplot(223)
ax3.plot([0,rot_matrix[0,0]], [0,rot_matrix[1,0]], 'r-')
ax3.plot([0,rot_matrix[0,1]], [0,rot_matrix[1,1]], 'b-')
ax3.plot([0,rot_vector[0]], [0,rot_vector[1]], 'g-')
ax3.set_xlim(-3,3)
ax3.set_ylim(-3,3)
ax3.set_title('Rotation Matrix')

ax4 = fig.add_subplot(224)
ax4.plot([0,com_matrix[0,0]], [0,com_matrix[1,0]], 'r-')
ax4.plot([0,com_matrix[0,1]], [0,com_matrix[1,1]], 'b-')
ax4.plot([0,com_vector[0]], [0,com_vector[1]], 'g-')
ax4.plot([0,shear_rot_vector[0]], [0,shear_rot_vector[1]], 'g-')

ax4.set_xlim(-3,3)
ax4.set_ylim(-3,3)
ax4.set_title('Composition Matrix')


plt.tight_layout()
fig = plt.gcf()
plt.show()
print('Original Matrix\n', orig_matrix, '\nShear Matrix\n', shear_matrix, '\nRotation Matrix\n', rot_matrix
        ,'\nComposition Matrix\n', com_matrix)
print('Original Vector\n', vector, '\nVector after Shear Transformation\n', shear_vector, '\nVector after Rotation Transformation\n', rot_vector
        ,'\nVector after Shear and Rotation Transformation\n', com_vector)
```
