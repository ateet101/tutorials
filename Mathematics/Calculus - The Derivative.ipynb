# The Derivative
Definition: The derivative is the slope of a tangent line at a small difference in x in a 2D plane.

Slope = $\frac{rise}{run}$

Lets define slope:

The graph below shows the distance that a car traveled in 10 second period:

```python
x = np.linspace(0,10,10)
y = np.array([0,0,5,15,40,80,100,110,115,115])
#(m,b) = polyfit(x,y,3)
z = polyfit(x,y,3)
#yp = polyval([m,b],x)
f = np.poly1d(z) 
print('Equation of the 3rd Degree Polynomial is given by the poly1d function\n', f)
x_new = np.linspace(0, 10, 50)
y_new = f(x_new)

plt.plot(x,y,'o')
plt.plot(x_new, y_new, label = 's(t)')
plt.plot([4,6],[f(4),f(4)], color = 'r', linestyle = '--', label = 'dt')
plt.plot([6,6],[f(4),f(6)], color = 'b', linestyle = '--', label = 'dy')
plt.ylabel('Distance (meters)')
plt.xlabel('Time (seconds)')
plt.title('Distance traveled by a car')
plt.legend()
plt.show()
```

The slope of are curve would be Slope = $\frac{rise}{run}$ or $\frac{\Delta y}{\Delta x}$


```python
x = np.linspace(0,10,10)
y = np.array([0,0,5,15,40,80,100,110,115,115])
#(m,b) = polyfit(x,y,3)
z = polyfit(x,y,1)
#yp = polyval([m,b],x)
f = np.poly1d(z) 
print('Equation of the line is given by the poly1d function\n', f)
x_new = np.linspace(0, 10, 50)
y_new = f(x_new)

#plt.plot(x,y,'o')
plt.plot(x_new, y_new, label = 's(t)')
plt.plot([4,4],[0,f(4)], color = 'r', linestyle = '--', label = 'a')
plt.plot([6,6],[0,f(6)], color = 'b', linestyle = '--', label = 'b')
plt.plot([0,4],[f(4),f(4)], color = 'r', linestyle = '--', label = 'f(a)')
plt.plot([0,6],[f(6),f(6)], color = 'b', linestyle = '--', label = 'f(b)')
plt.ylabel('Distance (meters)')
plt.xlabel('Time (seconds)')
plt.title('Distance traveled by a car')
plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

The slope of the line would be Slope = $\frac{rise}{run}$ or $\frac{\Delta y}{\Delta x}$
or $\frac{f(b)-f(a)}{b-a}$ 

```python
x = np.linspace(0,10,10)
y = x**2
#y = np.array([0,0,5,15,40,80,100,110,115,115])
#(m,b) = polyfit(x,y,3)
m,b = polyfit(x,y,1)
z1 = polyfit(x,y,2)
yp = polyval([m,b],x)
f1 = np.poly1d(z1)
print('Equation of the curve is given by the poly1d function\n', 'x^2')
x_new1 = np.linspace(0, 10, 50)
y_new1 = f1(x_new1)
plt.plot(x,y,'o')
#x=5
plt.plot(x, 5*x, label = 's(5)')
plt.plot(x_new1, y_new1, label = 's(t)')

plt.plot([5,5],[0,f1(5)], color = 'r', linestyle = '--', label = 'x')
plt.plot([6,6],[0,f1(6)], color = 'b', linestyle = '--', label = 'x+dt')
plt.plot([0,5],[f1(5),f1(5)], color = 'r', linestyle = '--', label = 'f(x)')
plt.plot([0,6],[f1(6),f1(6)], color = 'b', linestyle = '--', label = 'f(x+dt)')
plt.ylabel('Distance (meters)')
plt.xlabel('Time (seconds)')
plt.title('Distance traveled by a car')
plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

The slope of are tangent line wher S(t) crosses s(5) would be Slope = $\frac{rise}{run}$ or $\frac{\Delta y}{\Delta x}$
or $\frac{f(x+dt)-f(x)}{(x+dt)-x}$ 
= $\frac{f(x+dt)-f(x)}{x}$ 

As dt approaches 0, the blue line is approaching the red line. This is the slope of the line at a certain point $\Delta x = dt$

$$\lim_{dt\to0} \frac{f(x+dt)-f(x)}{dt}$$ 

### Formal Definition Derivation
$$y = x^2$$
$$\lim_{dt\to0} \frac{(x+dt)^2-x^2}{dt}$$ 
$$\lim_{dt\to0} \frac{x^2+2xdt+dt^2-x^2}{dt}$$ 
$$\lim_{dt\to0} 2x+dt = 2x$$
$$f(x)=Ax^{n}$$
$$f^1(x)=n*Ax^{n-1}$$
Slope of the tangent line around time = 5 seconds, which is shown by the line, is
$$f(5) = 2(5) = 10$$

## Finding Derivatives using Python:

1) import sympy library

```python
from sympy import *
```

2) Create symbolic variables

```python
x, y, z = symbols('x y z')
```

3) Create Equation $x^2$

```python
y = x**2
```

4) Take the derivative in terms of x $\frac{dy}{dx}$

```python
derivative = diff(y, x)
print(derivative)
```

5) Evaluate the derivative at a certain point x

```python
derivative.subs(x,3)
```

OR Solve for a list of x values using numpy and lambdify

```python
import numpy 
a = numpy.arange(10) 
derivative = diff(x**2, x)
f = lambdify(x, derivative, "numpy") 
f(a) 
```
