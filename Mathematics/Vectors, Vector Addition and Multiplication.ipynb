## Theory
Definition: A vector has a direction and a magnitude
The magnitude of a vector is defined by $||v|| = \sqrt{x^2 + y^2}$

Example: $v = [3,4]$ therefore $||v|| = \sqrt{3^2 + 4^2}=5$

Let's view a 1 Dimensional Vector with coordinates of $||v|| = [5,0]$ in Python

## 1 Dimensional Vector

```python
vector1 = np.array([5,0]) # v1 = [x1, y1] = [5,0]
plt.plot([0,vector1[0]],[0,vector1[1]])
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xlim(0,5)
plt.ylim(-1,5)
plt.title('1D Vector')
plt.show()
```

As seen from the plot, the line covers only 1 dimension in the x axis range. The vector starts at point (0,0) and points to (5,0).  _

Let's view a vector in 2D place pointing to {3,4} with the same origin as (0,0)

## Vector Space

$R_2$ stands Real Numbers in 2nd Dimensional Coordinate Space. A vector in 2D space is denoted by $v=[x_1,x_2]$

$R_3$ stands Real Numbers in 3rd Dimensional Coordinate Space. A vector in 3D space is denoted by $v=[x_1,x_2,x_3]$

$R_n$ stands Real Numbers in 'n' Dimensional Coordinate Space. A vector in nD space is denoted by $v=[x_1,x_2, x_3, x_4,....,x_n]$

## Vector Addition

Vector addition involves adding two vectors in the same dimension 
Let a, b, and c be vectors

$a + b = c$,
\begin{matrix}
\begin{bmatrix}
a_1\\
a_2 
\end{bmatrix}
+
\begin{bmatrix}
b_1\\
b_2 
\end{bmatrix}
=
\begin{bmatrix}
a_1+a_2\\
b_1+b_2
\end{bmatrix}
\end{matrix}

Example: 

\begin{matrix}
\begin{bmatrix}
6\\
-2 
\end{bmatrix}
+
\begin{bmatrix}
-4\\
4 
\end{bmatrix}
=
\begin{bmatrix}
2\\
2
\end{bmatrix}
\end{matrix}
Visual in Python:

```python
fig = plt.figure()
ax4 = fig.add_subplot(2,1,1)
ax5 = fig.add_subplot(2,1,2)

a = np.array([6,-2])
b = np.array([-4,4])
c = a + b

ax4.set_title('Vector Addition')
ax4.set_xlabel('x axis')
ax4.set_ylabel('y axis')
#ax4.set_xlim([0, 5])
#ax4.set_ylim([0, 5])
ax5.set_title('Vector Addition Tail to Head')
ax5.set_xlabel('x axis')
ax5.set_ylabel('y axis')


ax4.plot([0,a[0]],[0,a[1]], color = 'b', label = 'a')
ax4.plot([0,b[0]],[0,b[1]], color = 'r', label = 'b')
ax4.plot([0,c[0]],[0,c[1]], color = 'g', label = 'c = a+b')

ax5.plot([c[0],c[0]-a[0]],[c[1],c[1]-a[1]], color = 'b', label = 'a')
ax5.plot([0,b[0]],[0,b[1]], color = 'r', label = 'b')
ax5.plot([0,c[0]],[0,c[1]], color = 'g', label = 'c = a+ b')

fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

plt.legend()
plt.show()
```

## Vector Multiplication

Vector addition involves adding two vectors in the same dimension 
Let v be a vectors

$v=[v_1,v_2]$,

$3*v=[3*v_1,3*v_2]$,

Example: Let $a=[2,1]$

$2*a=2*[2,1]$

$a=[4,1]$

Visual in Python:

```python
v = np.array([[0, 2],[0, 1]]) #a = [2,1]
plt.xlim(-5,5)
plt.ylim(-5,5)
v2 = v *2 #a = [4,2]
v_neg = v*-2 #a = [-4,-2]
plt.title('Vector Multiplication')

plt.plot(v2[0],v2[1], color = 'r', label = 'a times 2')
plt.plot(v[0],v[1], color = 'b', label = 'a')

plt.plot(v_neg[0],v_neg[1], color = 'r', label = 'a times -2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```
