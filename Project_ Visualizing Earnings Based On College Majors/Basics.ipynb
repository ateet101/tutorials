```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads =  pd.read_csv('recent-grads.csv',delimiter=',')
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter',xlim=[0,2500])
ax.set_title('Median vs. Sample_size')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', xlim=[0,5000])
ax.set_title('Unemployment Rate vs. Sample_size')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter', xlim=[0,5000])
ax.set_title('Median vs. Full Time')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Share of Women vs. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter', xlim=[0,5000])
ax.set_title('Median vs. Men')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter', xlim=[0,5000])
ax.set_title('Women vs. Full Time')
```

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))

for  i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    ax = recent_grads[cols[i]].hist()
    
plt.show()
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))

for r in range(4,8):
    ax1 = fig.add_subplot(4,1,r-3)
    ax1 = recent_grads[cols[r]].plot(kind='hist', rot=40)
    
plt.show()
```
