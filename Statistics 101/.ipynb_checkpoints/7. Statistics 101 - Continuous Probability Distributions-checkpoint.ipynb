# I Uniform Probabiliy Distribution

Uniform probability of each discrete outcome takes the form:

Probability each outcome $= \frac{1}{n}$
where n is the number of outcomes

Data such as temperature, distance, income, mass, etc. that can be measured very precisely to several decimal points have INFINITY outcomes

Probability $= \frac{1}{\infty}$
where n = $\infty$ is the number of outcomes

$\frac{1}{\infty} = 0$

$$\lim_{n\to\infty} \frac{1}{n} = 0$$

A bar chart where the height of an infinite number of bars is...ZERO! We call this CONTINUOS data. The probability of any specific outcome is zero

Instead of the probability of any specific outcome (which is ZERO), for continuous data we can only find the probability over a SPECIFIED INTERVAL

Instead of probability of the roll of a dice $\frac{1}{2}$

we have, $f(x) = \frac{1}{b - a}$

\begin{equation}
f(x) = \frac{1}{b - a}
\end{equation}

\begin{equation}
ExpectedValue(x) = \frac{b + a}{2}
\end{equation}

\begin{equation}
Variance(x) = \sigma^2 = \frac{(b - a)^2}{12}
\end{equation}

\begin{equation}
StandardDeviation(x) = \sigma = \frac{(b - a)}{\sqrt{12}}
\end{equation}

## Problem Television Commercials
On average, 30-minute TV shows have 22 minutes of actual program. Let's assume the probability distribution for number of minutes of actual program is uniformly distributed from a low of 18 minutes to a high of 26 minutes.

a = 18, b = 26

\begin{equation}
f(x) = \frac{1}{26-18} = \frac{1}{8}
\end{equation}

\begin{equation}
ExpectedValue(x) = \frac{18 +26}{2} = 22
\end{equation}

\begin{equation}
Variance(x) = \sigma^2 = \frac{(26 - 18)^2}{12} = 5.33
\end{equation}

\begin{equation}
StandardDeviation(x) = \sigma = \frac{(26-18)}{\sqrt{12}} = 2.31
\end{equation}

Question: 

What is the probability P(x) the show will have at least 25 minutes of programming?

$x_1 = 25$, $x_2 = 26$
\begin{equation}
P(x) = \frac{x_2-x_1}{b-a} = \frac{26-25}{26-18} = .125
\end{equation}

Question: 

What is the probability P(x) the show will have between 21 and  25 minutes of programming?

$x_1 = 21$, $x_2 = 26$
\begin{equation}
P(x) = \frac{x_2-x_1}{b-a} = \frac{25-21}{26-18} = .5
\end{equation}

Question:

What is the probability P(x) the show will have between 22.32 and 24.77 minutes of programming?

$x_1 = 22.32$, $x_2 = 24.77$
\begin{equation}
P(x) = \frac{x_2-x_1}{b-a} = \frac{24.77-22.32}{26-18} = .306
\end{equation}

# Normal Distribution

Oftentimes data is described as being "normal"; but what does that mean?

Let's begin by talking about the frequency by which some events occur; both natural and man-made
- Natural: human height, temperature, blood-pressure, etc
- Man-made: machined products, financial data, sales, etc

For these measures, the average (mean) tends to be very frequent while measures away from the mean are less and less frequent

```python
print('Normal Distribution')
mu = 0
sigma = 2
x_normal = np.random.normal(mu, sigma, size=100)
sns.distplot(x_normal, fit=ss.norm, kde=False);
```

Properties:
- Lower Tail, Probabilty Area, Upper Tail are left to right
- The curve is symmetric
- The top of the curve is the mean, median, and mode
- The position is given by the mean
- The shape is determined by the Standard Deviation

```python
print('Normal Distribution')
x = np.linspace(-5, 5, 5000)
mu=0
sigma=.5
y = ss.norm.pdf(x, mu, sigma)
plt.plot(x,y)
y = ss.norm.pdf(x, 2, sigma)
plt.plot(x,y)
y = ss.norm.pdf(x, -2, sigma)
plt.plot(x,y)
plt.xlim(-6,6)
plt.ylim(0,1.5)
plt.title('mean is -2,0,2')
```

```python
print('Normal Distribution')
x = np.linspace(-5, 5, 5000)
mu=0
sigma=.5
y = ss.norm.pdf(x, mu, sigma)
plt.plot(x,y)
y = ss.norm.pdf(x, mu, .3)
plt.plot(x,y)
y = ss.norm.pdf(x, mu, .7)
plt.plot(x,y)
plt.xlim(-6,6)
plt.ylim(0,1.5)
plt.title('Standard Deviation is .3,.5,.7')
```

```python
print('Z Distribution')
x = np.linspace(-5, 5, 5000)
mu=0
sigma=1
y = ss.norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.xlim(-6,6)
plt.ylim(0,1.5)
plt.title('Standard Deviation is 1 and Mean is 0')
```

z = \frac{x-\mu}{\sigma}
\end{equation}

## T- Distribution
- When our sample size is n < 30 and/or we do not know the variance/std of the population we use t-distribution
- In the T distribution the bell curve is squished in the middle and fatter on the ends (tails)
- As n > 30 and definitely by n > 1000, the T distribution and standard normal Z distribution become indistinguishable

```python
mu = 0
k_values = [1E10, 10, 5, 1.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(-10, 10, 1000)
labels = [r'$\mathrm{t}(k=\infty)$', r'$\mathrm{t}(k=%.1f)$' % k_values[1], r'$\mathrm{t}(k=%.1f)$' % k_values[2], r'$\mathrm{t}(k=%.1f)$' % k_values[3]]
color = ['pink','yellow','green','blue']
#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(15, 7.75))

for k, ls, lb, color in zip(k_values, linestyles, labels, color):
    dist = ss.t(k, 0)

    if k >= 1E10:
        label = r'$\mathrm{t}(k=\infty)$'
    else:
        label = r'$\mathrm{t}(k=%.1f)$' % k

    plt.plot(x, dist.pdf(x), ls=ls, c=color, label=lb)

plt.xlim(-5, 5)
plt.ylim(0.0, 0.45)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|k)$')
plt.title("Student's $t$ Distribution")

plt.legend()
plt.show()
```
