# Confidence Interval Estimation, Sigma Known
Problem: Gumball Guessing Game

The object of the game is for you to guess how many gumballs are in the jar
- If you guess within 5, you get the gum plus $50.
- If you guess within 15, you get the gum plus $15.
- If you guess within 30, you get the gum plus

BUT beforehand, you have to decide which guess to make, How confident are you?

Problem: Checking Account Balance

A bank vice president is interested in the average checking account balance for all personal accounts in the affiliate region.

A random sample of 500 accounts is selected from the bank database to compute the average balance of all accounts.

Since this estimate is going to be used in company resorts, the sample mean must be a good approximation of all accounts.

What level of "confidence" for the mean is the VP going to be satisfied with? What +- interval number is going to be acceptable?

```python
## Confidence Intervals
mu=0
sigma= 1.1
x = np.linspace(-4, 4, 5000)
y = ss.norm.pdf(x, mu, sigma)
between = np.where(np.round(x) == -3)[0][0]
between1 = np.where(np.round(x) == 3)[0][-1]
fig, ax1 = plt.subplots(1, 1, sharex=True)
ax1.plot(x,y)
ax1.fill_between(x[between:between1], 0, y[between:between1])
ax1.set_title('The Standard Normal Curve')
ax1.set_ylim(0,.4)
ax1.text(-1.25,.15,'Probability under\n the curve = 1')
print('Probability under the curve = 1')
```

```python
## Confidence Intervals
between = np.where(np.round(x) == -2)[0][313]
between1 = np.where(np.round(x) == 2)[0][313]
color1 = np.where(np.round(x) == -3)[0][0]
color2 = np.where(np.round(x) == -2)[0][313]
color3 = np.where(np.round(x) == 2)[0][313]
color4 = np.where(np.round(x) == 3)[0][-1]
fig, ax1 = plt.subplots(1, 1, sharex=True)
ax1.plot(x,y)
ax1.fill_between(x[between:between1], 0, y[between:between1])
ax1.fill_between(x[color1:color2], 0, y[color1:color2], facecolor ='green')
ax1.fill_between(x[color3:color4], 0, y[color3:color4], facecolor ='green')
ax1.set_title('The 95% Probability Interval')
ax1.set_ylim(0,.4)
ax1.text(-1.25,.15,'95.4% of all sample\n means are in here')
ax1.text(-1.75,.05,'sample mean +- 1.96 STD')
ax1.text(-4.25,.020,'-1.96 STD')
ax1.text(3,.020,'1.96 STD')
print('95.44% of all sample means are in here')
```

The area in the tails is called alpha

a = 5% or .05

$\frac{a}{2} = \frac{.05}{2} = .025$

```python
## Confidence Intervals
between = np.where(np.round(x) == -3)[0][313]
between1 = np.where(np.round(x) == 3)[0][313]
color1 = np.where(np.round(x) == -3)[0][0]
color2 = np.where(np.round(x) == -3)[0][400]
color3 = np.where(np.round(x) == 3)[0][250]
color4 = np.where(np.round(x) == 3)[0][-1]
fig, ax1 = plt.subplots(1, 1, sharex=True)
ax1.plot(x,y)
ax1.fill_between(x[between:between1], 0, y[between:between1])
ax1.fill_between(x[color1:color2], 0, y[color1:color2], facecolor ='green')
ax1.fill_between(x[color3:color4], 0, y[color3:color4], facecolor ='green')
ax1.set_title('The 99% Probability Interval')
ax1.set_ylim(0,.4)
ax1.text(-1.25,.15,'99.7% of all sample\n means are in here')
ax1.text(-1.75,.05,'sample mean +- 2.96 STD')
ax1.text(-4.25,.020,'-2.96 STD')
ax1.text(3,.020,'2.96 STD')
print('99.7% of all sample means are in here')
```

### The standard deviation of the sampling distribution is called the standard error of the mean
- The $\sigma_{\bar x}$ is dependent on the SAMPLE SIZE
- If the sample size is small, what will happen to $\sigma_x$ and therefore the distribution? LARGER and WIDER
- $\sigma_{\bar x} = \frac{\sigma}{\sqrt{n}}$

## Interpretation
- The randomness lies in the elements chosen for the sample; NOT the population mean.
- The proportion of samples, size n, for which our estimate, the sample mean $\bar x$, is within a certain distance +- of the true population mean, $\mu$
- The sample mean is either within +- interval of the true mean, or it is not
- The confidence interval IS NOT  the probability that the population mean lies with the interval
