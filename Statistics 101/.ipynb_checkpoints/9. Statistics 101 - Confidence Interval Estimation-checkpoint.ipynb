### The standard deviation of the sampling distribution is called the standard error of the mean
- The $\sigma_{\bar x}$ is dependent on the SAMPLE SIZE
- If the sample size is small, what will happen to $\sigma_x$ and therefore the distribution? LARGER and WIDER
- $\sigma_{\bar x} = \frac{\sigma}{\sqrt{n}}$

## Interpretation
- The randomness lies in the elements chosen for the sample; NOT the population mean.
- The proportion of samples, size n, for which our estimate, the sample mean $\bar x$, is within a certain distance +- of the true population mean, $\mu$
- The sample mean is either within +- interval of the true mean, or it is not
- The confidence interval IS NOT  the probability that the population mean lies with the interval

## Example 1: Restaurant Sales
To Estimate then mean amount spent per customer at a national steakhouse chain, data was collectedbfor 75 customers. We are to assume the population standard deviation is $4

Questions:
1. At 95% confidence, what is the margin of error?
2. If the sample mean is $20, what is the 95% confidence interval for the population mean?

Answers:

1. 
\begin{equation}
\bar x \pm 1.96\sigma_{\bar x}\\
\sigma_{\bar x} = \frac{\sigma}{\sqrt n} = \frac{4}{\sqrt{75}} = 0.46\\
\bar x \pm 1.96*(0.46)\\
\bar x \pm .91\\
\end{equation}

2. 
\begin{equation}
\bar x \pm .91\\
20 \pm .91
\end{equation}

```python
mu=0
sigma= 1.1
x = np.linspace(-4, 4, 5000)
y = ss.norm.pdf(x, mu, sigma)
between = np.where(np.round(x) == -2)[0][313]
between1 = np.where(np.round(x) == 2)[0][313]
fig, ax1 = plt.subplots(1, 1, sharex=True)
ax1.plot(x,y)
ax1.fill_between(x[between:between1], 0, y[between:between1])
ax1.set_title('The 95% Probability Interval')
ax1.set_ylim(0,.4)
ax1.text(-.25,.15,'20 +- 0.91')
print('95.44% of all sample means are in here')
```

- All samples of n = 75 will have 0.91 as the margin of error, assuming $\sigma$ is known to be 4
- 95% of all intervals made using $\bar x \pm 0.91$ will contain the unknown population mean
- If we take 100 samples of n =75, and make intervals $\bar x \pm 0.91$, 95 of them will contain $\mu$

## Sample Standard Deviation
- When we do not know the population deviation, we use the sample standard deviation to approximation it.
- When $\sigma$ is unknown we estimate it with the sample standard deviation, s.
- Since every sample will have a unique s, samples of the same size do NOT necessarily have the same standard error

Standard Error will be :
Standard error of the mean
\begin{equation}
s_{\bar x} = \frac{s}{\sqrt n}
\end{equation}

## Interpretation
- The randomness lies in the elements chosen for the sample; NOT the population mean.
- It is the probability of obtaining a representative sample.
- The proportion of samples, size n, for which our estimate, the sample mean, $\bar x $, is within a certain distance $\pm$ of the true population mean, $\mu$
- The sample mean is either within $\pm$ interval of the true mean, or it is not
- The probability that the sample mean of sample size n is within a certain distance $\pm$ of the true population mean.

# Confidence Intervals, Estimating Sample Size Needed
## Margin of Error
$\bar x \pm \frac{Z_a}{2} \frac{\sigma}{\sqrt n}$

$E = \frac{Z_a}{2} \frac{\sigma}{\sqrt n}$

$\frac{Z_a}{2}$ = z-boundary of interval probability

$\frac{\sigma}{\sqrt n}$ = standard error

1. We choose E, the margin of error
2. We choose our confidence probability$\frac{Z_a}{2}$
3. We are given or we estimmate the standard deviation, $\sigma$
4. Solve for n

$n = \frac{(\frac{Z_a}{2})^2\sigma^2}{E^2}$

## A note about $\sigma$
Solving for the sample size requires the popluation standard deviation, $\sigma$. Most often we do not know it so we have to use an estimate or "planning value" in its place. Here are a few options:
1. Estimate $\sigma$ from previous studies using the same population of interest.
2. Conduct a pilot study to select a preliminary sample, Use the sample standard deviation from the pilot study.
3. Use a judgement or "best guess" for $\sigma$. A common guess is the data range (high - low) divided by 4

## Example 1
How large a sample should be selected to provide 95% confidence($\alpha$ = .05) interval with a margin of error (E) of 8? Assuming the population standard deviation, $\sigma = 36$

\begin{equation}
n = \frac{(\frac{Z_a}{2})^2\sigma^2}{E^2} \\
n = \frac{(1.96)^2*36^2}{8^2} = 77.8
\end{equation}

To have 95% of our sample means contain $\mu$, we need a sample size of 78

The question we are asking is "What minimum sample size is necessary to produce 95% confidence that the sample mean is $\pm$8 of the true population mean?"

As we increase sample size we reduce the standard error and our sample most likely becomes more representative of the population.

In graphical terms, we set the upper and lower boundary, Then we increase sample size, pulling the distribution up in the middle and inwards on the sides.

At some "sweet spot" for sample size, 95% of the probability is between our margin of error

## Example 2: Gas Problems
According to the Chicago Tribune website, the average price of gas inside the I-294 loop was 3.99. Based on preliminary analysis the standard deviation is .05

Recommend an appropriate sample size to determine a margin of error at 95% confidence
a. $\pm$ .05
b. $\pm$ .03
c. $\pm$ .01


\begin{equation}
n = \frac{(\frac{Z_a}{2})^2\sigma^2}{E^2} \\
n = \frac{(1.96)^2*.05^2}{.05^2} = 4 \\
n = \frac{(1.96)^2*.05^2}{.03^2} = 11 \\
n = \frac{(1.96)^2*.05^2}{.01^2} = 97
\end{equation}

As margin of error decreases, n increases
