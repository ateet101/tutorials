## Multiple Regression
Multiple regression is an extension of simple linear regression.

Simple linear regression is a one-to-one relationship - I.V -> D.V.

Multiple regression is a many-to-one relationship - 2 or more I.V -> D.V.

## New Considerations
- Adding more independent variables to a multiple regression procedure does not mean the regression will be "better" or offer better predictions; in fact it can make things worse. This is called OVERFITTING
- The addition of more independent variables creates more relationships among them. So not only are the independent variables potentially related to the dependent variable, they are also potentially related to each other. When this happens, it is called MULTICOLLINEARITY
- The ideal is for all of the independent variables to be correlated with the dependent variable but NOT with each other
- Because of multicollinearity and overfitting, there is a fair amount of prep work to do before conducting multiple regression analysis if one is to do it properly
-- Correlations
-- Scatter Plots
-- Simple regression

## More Relationships
- Independent Variable: -- milesTraveled and numDeliveries
- Dependent Variable: -- travelTime

This also creates a relationship between the two indepedent variables which causes Potential multicollinearity

If the two independent variables are related to each other, we don't know which causes the variablity in the dependent variable

## Multiple Regression Model
Multiple Regression Model
\begin{equation}
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... \beta_px_p + \epsilon
\end{equation}
$\beta_0 + \beta_1x_1 + \beta_2x_2 + ... \beta_px_p$ are the linear parameters
$\epsilon$ error

Multiple Regression Equation
\begin{equation}
E(y) = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... \beta_px_p
\end{equation}
$\epsilon$, error term assume to be 0

Estimated Multiple Regression Equation
\begin{equation}
\hat y = b_0 + b_1x_1 + b_2x_2 + ... b_px_p
\end{equation}
$b_0, b_1, b_2, ... b_p$ are the estimates of $\beta_0, \beta_1, \beta_2, ... \beta_p$ 

$\hat y$ = preditcted value of the dependent variable

## Estimated Multiple Regression Equation  
#### Example
$\hat y = 27 + 9x_1 + 12x_2$
- $x_1$ = captial investmenst (\$1000s)
- $x_2$ = marketing expenditures (\$1000s)
- $\hat y$ = predicted sales

In multiple regression, each coefficient is interpreted as the estimated change in y corresponding to a one unit change in a variable, when all other variables are held constant

So in this example, \$9000 is an example of the expected increase in sales y, corresponding to a \$1000 increase in capital investments $(x_1)$ when marketing expenditures $(x_2)$ are held constant

## Review
- Multiple regression is an extension of simple linear regression
- Two or more independent variables are used to predict/explain the variance in one dependent variable
- Two problems may arise: overfitting and multicollinearity
- In multiple regression, each coefficient is interpreted as the estimated change in y corresponding to a one unit change in a variable, when all other variables are held constant
#### Overfitting 
is caused by adding too many independent variables; they account for more variance but add nothing to the model
##### Multicollinearity 
happens when some/all of the independent variables are correlated with each other

#### VIF
One way to measure multicollinearity is the variance inflation factor (VIF), which assesses how much the variance of an estimated regression coefficient increases if your predictors are correlated.  If no factors are correlated, the VIFs will all be 1.

If the VIF is equal to 1 there is no multicollinearity among factors, but if the VIF is greater than 1, the predictors may be moderately correlated. The output above shows that the VIF for the Publication and Years factors are about 1.5, which indicates some correlation, but not enough to be overly concerned about. A VIF between 5 and 10 indicates high correlation that may be problematic. And if the VIF goes above 10, you can assume that the regression coefficients are poorly estimated due to multicollinearity.

## Example RE/Modelers Inc
You are an analyst for a small company that develops house pricing models for independent realtors. To generate your models you use publicly available data such as; list price; square footage, number of bedrooms, number of bathrooms, ect

You are interested in another question: Is the public high school in the neighborhood "Exemplary" (highest rating) and how is that rating related to home price?

High school rating is not quantitative, it is qualitative (categorical). For each home the high school is either exemplary or not; yes or no

## Home Price Data

```python
e = 'Exemplary'
ne = 'Not exemplary'
home_df = pd.DataFrame({'price1000s(y)': [145,69.9,315,144.9,134.9,369,95,228.9,149,295,388.5,75,130,174,334.9], 
                        'sqft(x1)': [1872,1954,4104,1524,1279,3278,1192,2252,1620,2466,3188,1061,1195,1552,2901], 
                        'exempHS(x2)':[ne,ne,e,ne,ne,e,ne,e,ne,e,e,ne,ne,e,e]})
home_df[['price1000s(y)','sqft(x1)','exempHS(x2)']]
```

```python
home_df['exempHS(x2)'] = home_df['exempHS(x2)'].apply(lambda x: 0 if x == 'Not exemplary' else 1)
home_df[['price1000s(y)','sqft(x1)','exempHS(x2)']]
```

```python
#home_df.plot.scatter(x = 'sqft(x1)', y = 'price1000s(y)', hue = 'exempHS(x2)')
'''
groups = home_df.groupby('exempHS(x2)')

# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group['sqft(x1)'], group['price1000s(y)'], marker='o', linestyle='', ms=12, label=name)
ax.legend()
ax.set_title()
'''
sns.pairplot(x_vars=['sqft(x1)'], y_vars=['price1000s(y)'], data=home_df, hue='exempHS(x2)', size=5)
```

## Interpretation
Multiple Regression Equation
\begin{equation}
E(y) = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... \beta_px_p
\end{equation}

Expected value of home price given the high school is NOT exemplary, $x_2 =0$

E(y|Not Exemplary) = $\beta_0 + \beta_1x_1 + \beta_2(0)$

E(y|Not Exemplary) = $\beta_0 + \beta_1x_1$

Expected value of home price given the high school is exemplary, $x_1 =1$

E(y|Exemplary) = $\beta_0 + \beta_1x_1 + \beta_2(1)$

E(y|Exemplary) = $\beta_0 + \beta_1x_1 + \beta_2$

E(y|Exemplary) = $(\beta_0 + \beta_2) + \beta_1x_1$

```python
df = pd.DataFrame({'price1000sY': home_df['price1000s(y)'], 'sqftX1': home_df['sqft(x1)'], 'exempHSX2': home_df['exempHS(x2)']})
model = smf.ols('price1000sY ~ sqftX1 + exempHSX2',df).fit()
print(model.summary())
print(anov.anova_lm(model))
```

High R-sq(85.7%) and R-sq(adj)(83.31%) along with a R-sq(pred)(70.3%) that doesn't fall off a cliff are all food signs

Sqft: Every square foot is related to an increase in price of 0.0621 (\$1000s) or \$62.10 per square foot

ExempHS: On average, a home in an area with an exemplary high schooi is related to a $98,600 higher price

```python
print('Estimated Regression Equation for Re/Model, Inc \n E(y) = {} + {}x1 + {}x2'.format(model.params[0],model.params[1],model.params[2]))
```

```python
print('Expected value of home price given the HS is NOT exemplary, x2 = 0 \nE(y|Not exemplary) = {} + {}x1 + {}(0)'.format(model.params[0],model.params[1],model.params[2]))
print('E(y|Not exemplary) = {} + {}x1'.format(model.params[0],model.params[1]))

print('\n\nExpected value of home price given the HS IS exemplary, x2 = 1 \nE(y|Exemplary) = {} + {}x1 + {}(1)'.format(model.params[0],model.params[1],model.params[2]))
print('E(y|Exemplary) = {} + {}x1 + {}'.format(model.params[0],model.params[1],model.params[2]))
print('E(y|Exemplary) = ({} + {}) + {}x1'.format(model.params[0],model.params[1],model.params[2]))
print('E(y|Exemplary) = {} + {}x1'.format(model.params[0]+model.params[2],model.params[1]))
```
