```python
def chi(n, sample_variance, pop_variance):
    chi_squared = (((n-1)*sample_variance)/pop_variance)
    return chi_squared

def population_variance(n, sample_variance, chi_squared):
    pop_variance = []
    for chi in chi_squared:
            pop_variance.append(((n-1)*sample_variance)/chi)
    return pop_variance
```

## Variance Review
Variance and its square-root, the standard deviation, are both measures of the spread or variability in data. Two or more data sets could have the same mean, but very differnt variances.

Population Variance
\begin{equation}
\sigma^2 = \frac{\sum (\bar x_i - \mu)^2}{N}
\end{equation}

Sample Variance
\begin{equation}
s^2 = \frac{\sum (\bar x_i - \bar x)^2}{n-1}
\end{equation}

These are based on the squared difference between each observation and the mean. The average squared deviation. How far is each data point from the center (mean) on average?

## Sampling Distribution of $\bar x$
When we take many samples of the same size from a population and then find the sample means, $\bar x$, those sample means follow the normal curve when placed in their own distribution

## Sampling Distribution of $\bar s^2$
When we take many samples of the same size from a normal population and then find those sample variances, $s^2$, those sample variances DO NOT follow the normal curve when placed in their own distribution.

They follow the chi-square distribution, $\it X^2$, with n-1 degrees of freedom


```python
x = np.linspace(0, 30, 1000)
fig,ax = plt.subplots(1,1)

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [3,5,10,20]
for df, ls in zip(deg_of_freedom, linestyles):
  ax.plot(x, ss.chi2.pdf(x, df), linestyle=ls, label=r'$df=%i$' % df)

plt.xlim(0, 5)
plt.ylim(0, 0.5)

plt.xlabel('$\chi^2$')
plt.ylabel(r'$f(\chi^2)$')
plt.title(r'$\chi^2\ \mathrm{Distribution}$')
plt.ylim(0,.3)
plt.xlim(0,30)
plt.legend()
plt.show()
```

## Anatomy of the $\it X^2$ Distribution
1. There is no "one" chi-square distribution. There is a chi-square distribution for each n-1 degrees of freedom
2. Area (probability) under the curve is 1
3. The curve is asymptotic; it never touches the x-axis
4. In the chi-square, "1" is at the left side and "0" is on the right side
5. Therefore the cumulative probability runs right to left
6. The probabilities are found in the chi-square table in the same manner as normal curves

```python
x = np.linspace(0, 30, 1000)
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,2,1)
y = ss.chi2.pdf(x, 9)
linestyles = [':', '--', '-.', '-']
ax.plot(x, y, linestyle='-', label=r'$df=%i$' % 9)

ax.set_xlim(0, 5)
ax.set_ylim(0, 0.5)
ax.set_xlabel('$\chi^2$')
ax.set_ylabel(r'$f(\chi^2)$')
ax.set_title(r'$Chi-square\  \chi^2\ \mathrm{Distribution}$')
ax.set_ylim(0,.3)
ax.set_xlim(0,30)
ax.fill_between(x, y, where=(x>2.7) & (x<19.023))
ax.text(2.2, .05, '2.7')
ax.text(19.1, .03, '19.023')
ax.text(0.2, .03, '2.5%')
ax.text(25.1, .02, '2.5%')
ax.text(5.1, .02, '95%')
ax.text(20, .15, r'$\chi^2_{.975} \leq \chi^2 \leq \chi^2_{.025}$')
ax.text(20, .1, r'$2.7 \leq \chi^2 \leq 19.023$')
ax.text(1, -.01, r'$\chi^2_{.975}$')
ax.text(21, -.01, r'$\chi^2_{.025}$')

ax.legend()
```

Sampling Distribution
We know that the sampling distribution of variance follows the chi-square distribution, $\chi^2$. But how does that relate to the actual data we collect?

Whenever a random sample of size n is selected from a normal population, the sampling distribution of:
\begin{equation}
\chi^2 = \frac{(n-1)s^2}{\sigma^2}
\end{equation}

n = sample size

$s^2$ = sample variance

$\sigma^2$ = population variance (unknown)

df = n-1

## Interval Estimate of the Variance

\begin{equation}
\chi^2_{.975} \leq \chi^2 \leq \chi^2_{.025} \\
\chi^2 = \frac{(n-1)s^2}{\sigma^2} \\
\sigma^2\chi^2_{.975} \leq (n-1)s^2 \leq \sigma^2\chi^2_{.025} \\
\end{equation}

Split
\begin{equation}
\sigma^2\chi^2_{.975} \leq (n-1)s^2  \\
\sigma^2 \leq \frac{(n-1)s^2}{\chi^2_{.975}} \\
(n-1)s^2 \leq \sigma^2\chi^2_{.025} \\
\sigma^2 \geq \frac{(n-1)s^2}{\chi^2_{.025}} \\
\end{equation}

Recombine
\begin{equation}
\frac{(n-1)s^2}{\chi^2_{.975}} \leq 
\sigma^2 \leq \frac{(n-1)s^2}{\chi^2_{.025}} \\
\end{equation}

## Interval Estimate of the Variance

\begin{equation}
\chi^2_{.975} \leq \chi^2 \leq \chi^2_{.025} \\
3.82 \leq \chi^2 21.92 \\
\end{equation}

# Hypothesis Tests for the Variance 
## Example: High Interest Realty
High Interest Realty has set a timed sales goal whereby the standard deviation of listing -to-sell time is no longer than 21 days.

Since the standard deviation of 21 is the square root of the variance, $21^2 = 441$ is the variance upper limit. Realizing this goal will help with sales forecasting and marketing.

To conduct this analysis, the sales records for 15 previously sold homes were randomly selected. Based on this sample the following information was obtained:

\begin{equation}
n = 15 \\
\bar x = 162 days \\
s^2 = 582 \\
s = 24 days
\end{equation}

## Hypothesis Testing for Variance

```python
x = np.linspace(0, 30, 1000)
y = ss.chi2.pdf(x, 10)
fig1 = plt.figure(figsize=(20,5))
ax4 = fig1.add_subplot(1,3,1)
ax5 = fig1.add_subplot(1,3,2)
ax6 = fig1.add_subplot(1,3,3)
ax4.plot(x, y )
ax4.fill_between(x, y, where=(x>2.7) & (x<19.023))
ax4.set_title(r'$H_O: \sigma^2 = \sigma_O^2$' + '\n' + r'$H_a: \sigma^2 \neq \sigma_O^2$')
ax5.plot(x, y )
ax5.fill_between(x, y, where=(x<19.023))
ax5.set_title(r'$H_O: \sigma^2 \leq \sigma_O^2$' + '\n' + r'$H_a: \sigma^2 > \sigma_O^2$')
ax6.plot(x, y )
ax6.fill_between(x, y, where=(x>2.7))
ax6.set_title(r'$H_O: \sigma^2 \geq \sigma_O^2$' + '\n' + r'$H_a: \sigma^2 < \sigma_O^2$')
```

## High Interest Reality
Step 1: Establish Hypothesis

$H_O$: $\sigma \geq 441$
$H_a$: $\sigma > 441$

Step 2: Determine Appropriate Statistical Test and Sampling Distribution

This will be a one-tailed test. Since the concern is a larger than desired variance, it will be an upper/right tailed test.

We will use the chi-square distribution with 15-1 = 14df

\begin{equation}
\chi^2 = \frac{(n-1)s^2}{\sigma^2}
\end{equation}


Step 3: Specify the Type I error rate (significance level)

$\alpha = .05$

Step 4: State the decision rule

If $\chi^2 > 23.685$, reject $H_O$ per the chi-square table

Step 5: Gather data

$n = 15, s^2 = 582$

Step 6: Calculate test statistic

\begin{equation}
\chi^2 = \frac{(n-1)s^2}{\sigma^2} \\
\chi^2 = \frac{(15-1)582}{441} \\
\chi^2 = 14.48
\end{equation}


Step 7 & 8: State statistical conclusion


```python
x = np.linspace(0, 30, 1000)
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
y = ss.chi2.pdf(x, 14)
linestyles = [':', '--', '-.', '-']
ax.plot(x, y, linestyle='-', label=r'$df=%i$' % 11)

ax.set_xlim(0, 5)
ax.set_ylim(0, 0.5)
ax.set_xlabel('$\chi^2$')
ax.set_ylabel(r'$f(\chi^2)$')
ax.set_ylim(0,.1)
ax.set_xlim(0,30)
ax.fill_between(x, y, where=(x<23.685))
ax.text(24.1, .03, r'$\chi^2_{crit} = 23.685$')
ax.text(12, .03, r'$\chi^2 = 18.48$', color ='r')
ax.text(9.1, .02, '95%')
ax.text(20, .09, r'$H_O: \sigma^2 \leq 441$')
ax.text(20, .08, r'$H_a: \sigma^2 > 441$')
ax.axvline(19.48, 0.01, 0.03, color='r', linestyle=':', linewidth =10.5)

ax.legend()
```

Conclusion:
    
Fail to reject the null hypothesis.

The sample data does not offer sufficient evidence to conclude the list-to-close time variance exceeds the upper limit set by HIR

## Example: Micro Machines
A new precision machine tool is being installed in a local production plant. Based on laser-read measurements, the machine being replaced produced a tooling variance of $s^2 = 100$ microns

After a break-in period the new machine is tested to see if its production variance is the same as the previous machine.

Thirty samples are taken from the machine producing a sample variance, $s^2 = 162$ microns

Step 1: Establish Hypothesis

$H_O$: $\sigma = 100$
$H_a$: $\sigma \neq 100$

Step 2: Determine Appropriate Statistical Test and Sampling Distribution

This will be a two-tailed test. 

We will use the chi-square distribution with 30-1 = 29df

\begin{equation}
\chi^2 = \frac{(n-1)s^2}{\sigma^2}
\end{equation}


Step 3: Specify the Type I error rate (significance level)

$\alpha = .05$

Step 4: State the decision rule

If $\chi^2 > 16.047$, reject $H_O$ per the chi-square table

If $\chi^2 < 45.722$, reject $H_O$ per the chi-square table

Step 5: Gather data

$n = 30, s^2 = 162$

Step 6: Calculate test statistic

\begin{equation}
\chi^2 = \frac{(n-1)s^2}{\sigma^2} \\
\chi^2 = \frac{(30-1)162}{100} \\
\chi^2 = 46.98
\end{equation}


Step 7 & 8: State statistical conclusion


```python
x = np.linspace(0, 60, 1000)
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
y = ss.chi2.pdf(x, 29)
linestyles = [':', '--', '-.', '-']
ax.plot(x, y, linestyle='-', label=r'$df=%i$' % 29)

ax.set_xlim(0, 5)
ax.set_ylim(0, 0.5)
ax.set_xlabel('$\chi^2$')
ax.set_ylabel(r'$f(\chi^2)$')
ax.set_ylim(0,.1)
ax.set_xlim(0,90)
ax.fill_between(x, y, where=(x<45.722) & (x>16.047))
ax.text(40, .03, r'$\chi^2_{crit} = 45.722$')
ax.text(12, .03, r'$\chi^2_{crit} = 16.047$')

ax.text(49, .03, r'$\chi^2 = 46.98$', color = 'r')
ax.text(9.1, .02, '95%')
ax.text(20, .09, r'$H_O: \sigma^2 = 100$')
ax.text(20, .08, r'$H_a: \sigma^2 \neq 100$')
ax.axvline(46.98, 0.01, 0.03, color='r', linestyle=':', linewidth =10.5)

ax.legend()
```

Conclusion:
    
Reject the null hypothesis.

The sample data does offer sufficient evidence to conclude the production variance of the new machine is higher than that of the previous machine. Recalibration/repair of the new machine may be necessary.

# F-Ratio Test for Two Equal Variances
## Example: Terrific Tuna
The Terrific Tuna Company uses two machines to fill each 5 ounce can of tuna. The quality assurance manager wishes to compare the variability, the variances, of the two canning machines

$\sigma_x^2 = \sigma_y^2$?

To do so, a sample of cans is selected from each machine for testing

\begin{equation}
n_x = 25 \\
\bar x_x = 5.0592 oz. \\
s_x^2 = 0.1130 \\
s_x = 0.3361
\end{equation}

\begin{equation}
n_y = 22 \\
\bar x_y = 4.9808 oz. \\
s_y^2 = 0.0137 \\
s_y = 0.1171
\end{equation}

The easiest way to compare the relative size of two measurements is by using the F-ratio

$\frac{s_x^2}{s_y^2}$

$s_x^2$ = larger sample variance,
$s_y^2$ = smaller sample variance

$F = \frac{s_x^2}{s_y^2}$

## The F-Distribution
When independent random samples, $n_x$ and $n_y$ are taken from two normal populations with equal variances, the sampling distribution of the ratio of those sample variances follows the F-distribution:

\begin{equation}
F = \frac{s_x^2}{s_y^2}
\end{equation}

Python Code for finding critical values
